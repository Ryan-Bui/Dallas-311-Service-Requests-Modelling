```python
min_count = 1000  # threshold (you can adjust)

dept_counts = data_311['Department'].value_counts()

common_depts = dept_counts[dept_counts >= min_count].index

data_311['Department_grouped'] = data_311['Department'].where(
    data_311['Department'].isin(common_depts),
    'Other'
)
```

```python
data_311['Department_grouped'].value_counts()
```

```python
sample_frac = 0.1  # 10% sample

sampled_df = data_311.groupby('Department_grouped', group_keys=False)\
    .sample(frac=sample_frac, random_state=42)\
    .reset_index(drop=True)
```

```python
sampled_df
```

```python
original = data_311['Department_grouped'].value_counts(normalize=True)
sampled = sampled_df['Department_grouped'].value_counts(normalize=True)

comparison = pd.concat([original, sampled], axis=1)
comparison.columns = ['Original', 'Sampled']

print(comparison)
```

```python
sampled_df['Created Date'] = pd.to_datetime(sampled_df['Created Date'])
sampled_df['Closed Date'] = pd.to_datetime(sampled_df['Closed Date'])
```

```python
sampled_df['days_to_close'] = (
    sampled_df['Closed Date'] - sampled_df['Created Date']
).dt.total_seconds() / 3600  # convert to hours
```

```python
sampled_df
```

```python
cols_to_drop = [
    'Closed Date',      # leakage
    'Update Date',      # leakage
    'Status',           # leakage
    'Outcome',          # borderline leakage
    'Lat_Long Location' # messy/high precision
]

sampled_df = sampled_df.drop(columns=cols_to_drop)
```

```python
sampled_df['month'] = sampled_df['Created Date'].dt.month
sampled_df['day_of_week'] = sampled_df['Created Date'].dt.dayofweek
sampled_df['hour'] = sampled_df['Created Date'].dt.hour
```

```python
sampled_df = sampled_df.drop(columns=['Created Date'])
```

```python
sampled_df = sampled_df.drop(columns=['Department'])
```

```python
sampled_df['service_type_encoded'] = sampled_df.groupby(
    'Service Request Type'
)['days_to_close'].transform('mean')

sampled_df = sampled_df.drop(columns=['Service Request Type'])
```

```python
sampled_df['ERT_days'] = sampled_df['ERT (Estimated Response Time)'].str.extract('(\d+)').astype(float)

sampled_df = sampled_df.drop(columns=['ERT (Estimated Response Time)'])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
sampled_df['Priority'] = le.fit_transform(sampled_df['Priority'].astype(str))
```

```python
sampled_df['Method Received Description'] = le.fit_transform(
    sampled_df['Method Received Description'].astype(str)
)
```

```python
sampled_df['Department_grouped'] = le.fit_transform(
    sampled_df['Department_grouped'].astype(str)
)
```
